## Exercício - revisão
"""
    Qual é sua a altura?
    Qual é seu peso?
    Qual o seu IMC?

    função (mundo é baseado em funções)
    IMC = peso/altura²
    
"""

```python
altura = float(input("Qual é sua altura: ").replace(",","."))
peso = float(input("Qual é seu peso: "))

imc = peso / (altura ** 2)
print(f"seu IMC é: {imc:.2f}")

if imc < 18.5:
    print("Abaixo do peso")
elif 18.5 <= imc < 24.9:
    print("Peso normal")
elif 25 <= imc < 29.9:
    print("Sobrepeso")
elif 30 <= imc < 34.9:
    print("Obesidade Grau 1")
elif 35 <= imc < 39.9:
    print("Obesidade Grau 2")
else:
    print("Obesidade Grau 3 (Obesidade Mórbida)")
```

# Gráfico de linha

```python
import matplotlib.pyplot as plt
```

```python
from matplotlib.ticker import MultipleLocator
```

```python
plt.plot(dados_brasil['ano'],dados_brasil['imigrantes'], color='green')
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.yticks([500,1000,1500,2000,2500,3000])
plt.title('Gráfico de imigração de Brasileiros para o Canadá')
plt.grid()
plt.show()
```

# Transformando um gráficoem figura

```python
plt.figure(figsize=(8,5))
# 1 polegada = 2.54 cm
# largura 8 = 20.32 cm
# altura 4 = 10.16 cm
# PDI (dots per inch, ou pontos por polegada)
# Padrão no matplolib - DPI = 100
# Largura : 8 polegadas x 100 DPI = 800 pixels
# Altura : 4 polegadas x 100 DPI = 400 pixels
plt.plot(dados_brasil['ano'],dados_brasil['imigrantes'], color='green',markersize=3, )
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.yticks([500,1000,1500,2000,2500,3000])
plt.title('Gráfico de imigração de Brasileiros para o Canadá')
plt.annotate('Pico de imigração', xy=(29, 2600), xytext=(15, 1500),
             arrowprops=dict(facecolor='red', shrink=0.20))
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['ano'],dados_brasil['imigrantes'], color='green',markersize=3, marker='o',markerfacecolor='white' )
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.yticks([500,1000,1500,2000,2500,3000])
plt.title('Gráfico de imigração de Brasileiros para o Canadá', color='green')
plt.xlabel('Ano',fontsize=14)
plt.ylabel('imigrantes',fontsize=16)
plt.annotate('(Pico de imigração)', xy=(28, 2760), xytext=(40, 1700),
             arrowprops=dict(facecolor='red', shrink=0.20))
plt.fill_between(dados_brasil['ano'], dados_brasil['imigrantes'], color='darkgrey', alpha=0.7)

plt.grid()
plt.show()
```

```python
# Criando um 
fig,axs = plt.subplots(1,2,figsize=(15,6))

# Subplot 1: Gráfico de linha
axs[0].plot(dados_brasil['ano'],dados_brasil['imigrantes'], color='red', marker='o', markersize=3,markerfacecolor='white')
axs[0].xaxis.set_major_locator(MultipleLocator(5))
axs[0].set_title("Imigração do Brasil para o Canadá\nde 1980 a 2013")
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de imigrantes')
axs[0].fill_between(dados_brasil['ano'], dados_brasil['imigrantes'], color='darkgrey', alpha=0.7)
axs[0].grid()

# Subplot 2: Gráfico Boxplot
axs[1].violinplot(dados_brasil['imigrantes'])
axs[1].set_title("Violinplot da imigração do Brasil para o Canadá de \nde 1980 a 2013")
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de imigrantes')
axs[1].grid()


plt.show()
```
