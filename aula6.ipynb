# Temas Seaborn

```python
top_10 = df.sort_values(by='Total', ascending=False).head(10)
top_10
```

```python
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
```

```python
sns.barplot(data=top_10, x='Total', y=top_10.index)
plt.show()
```

```python
ax = sns.barplot(data=top_10, x='Total', y=top_10.index)
ax.set(
    title='Os dez paises com mais imigrantes no Canada',
    xlabel='Total de imigrantes',
    ylabel=''
)
plt.show()
```

```python
def gerar_grafico_paleta(pallete):
  fig,ax = plt.subplots(figsize=(15,6))
  ax = sns.barplot( data=top_10, y=top_10.index, x='Total', orient='h', palette=pallete)
  ax.set_title('Os dez paises com mais imigrantes no Canada', loc='left', fontsize=16)
  ax.set_xlabel('Total de imigrantes',fontsize=14)
  ax.set_ylabel('')

  for p in ax.patches:
    width = p.get_width()
    ax.text(width + 10000,
            p.get_y() + p.get_height() / 2,
            '{:1.0f}'.format(width),
            ha = 'left',
            va = 'center')

  plt.show()
```

```python
sns.set_theme(style='whitegrid')
gerar_grafico_paleta('tab10')
```

# Gráficos com python

```python
import plotly.express as px
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
brasil = df.loc['Brasil', anos]
brasil.head(10)
```

```python
dados_brasil = pd.DataFrame(brasil)
```

```python
brasil_dict = {'ano':brasil.index.tolist(), 'imigrantes':brasil.values.tolist()}
```

```python
dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil.head()
```

```python
fig = px.line(dados_brasil, x='ano', y='imigrantes', title='Imigrantes no Brasil para o Canadá de 1980 a 2013')
fig.update_layout(width=1000,height=500)
fig.show()
```

# Desafio os xaxis - tickangle

```python
fig = px.line(dados_brasil, x='ano', y='imigrantes', title='Imigrantes no Brasil para o Canadá de 1980 a 2013')
fig.update_layout(width=1000,
                  height=500,
                  font_family='Arial',
                  font_color='black',
                  title_font_color= 'black'

                  )
fig.update_traces(line_color='orange', line_width=3)
fig.update_xaxes(tickangle=-45)
fig.show()
```

```python
fig.write_html('imigrante_brasil.html')
```
